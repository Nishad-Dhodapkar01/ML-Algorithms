K Means Clustering:

Introduction:
K-Means Clustering is a simple, unsupervised machine learning algorithm used for grouping similar data points into clusters. The goal of K-Means is to partition the dataset into K distinct clusters, where each data point belongs to the cluster whose mean (average) is the closest.
It is called "K-Means" because:
"K" refers to the number of clusters you want to form.
"Means" refers to the average value of the data points in each cluster.


Definition:
K-Means is a centroid-based clustering algorithm. It works by finding K "centroids," which are the center points of each cluster. Then, it assigns each data point to the nearest centroid. The centroids are updated iteratively until the clusters stabilize (i.e., the centroids no longer change).


Steps of the K-Means Algorithm:
1. Choose K (Number of Clusters):
Decide how many clusters (K) you want to divide the data into. This is a hyperparameter that you need to specify before running the algorithm.

2. Initialize the Centroids:
Select K data points randomly from the dataset to serve as the initial centroids of the clusters.

3. Assign Data Points to the Nearest Centroid:
For each data point, calculate the distance (usually Euclidean distance) from the point to each centroid.
Assign the data point to the closest centroid, meaning the cluster with the nearest centroid.

4. Recalculate the Centroids:
After assigning all data points to a cluster, calculate the mean (average) of all data points in each cluster.
Update the centroid to this new mean position.

5. Repeat Steps 3 and 4:
Repeat the process of assigning points to the closest centroid and recalculating the centroids until the centroids stop changing (i.e., convergence) or a certain number of iterations is reached.

6. Output the Final Clusters:
Once the centroids have stabilized, the algorithm finishes. The data points are grouped into K clusters, and each data point belongs to the cluster with the nearest centroid.

```python
# Step 1: Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
# Step 2: Load the dataset (Iris dataset in this case)
iris = load_iris()
X = iris.data  # Features: Sepal length, sepal width, petal length, petal width
y = iris.target  # Target (not used in unsupervised learning)
```

```python
# Step 3: Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)  # Specify the number of clusters (3 for Iris dataset)
kmeans.fit(X)  # Fit the model
```

```python
# Step 4: Get the cluster labels
y_kmeans = kmeans.predict(X)  # Predicted clusters
```

```python
# Step 5: Plot the clusters (using PCA for dimensionality reduction to 2D)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_kmeans, cmap='viridis', marker='o', edgecolor='k', s=100, alpha=0.6)
```

```python
# Plot the cluster centroids
centroids = kmeans.cluster_centers_
centroids_pca = pca.transform(centroids)
plt.scatter(centroids_pca[:, 0], centroids_pca[:, 1], c='red', s=200, marker='X', label='Centroids')

plt.title("K-Means Clustering on Iris Dataset")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend()
plt.show()
```
